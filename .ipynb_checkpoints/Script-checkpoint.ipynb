Now , since we have close to 9 numerical features,instead of rewritting code to plot the distribution,we define a function and invoke the function everytime

```python
def plot_numerical(variable):
    plt.figure(figsize=(16,6))
    plt.subplot(121)
    sns.distplot(train[variable].values, bins=50, kde=False, color="red")
    plt.title(f'Histogram of {variable} in train')
    plt.xlabel(f'{variable}', fontsize=12)
    plt.subplot(122)
    sns.distplot(test[variable].values, bins=50, kde=False, color="red")
    plt.title(f'Histogram of {variable} in test')
    plt.xlabel(f'{variable}', fontsize=12)
    plt.show()
    
def plot_boxplot(variable):
    #print(f'Plotting boxplot for {variable}\n')
    plt.figure(figsize=(8,8))
    sns.boxplot(train['Cover_Type'],train[variable])
    plt.title(f'Boxplot of {variable} with CoverType in train dataset')
    plt.xlabel("Cover Type")
    
```

### Baseline Model

```python
##importing necessary libraries:
import lightgbm as lgb
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split,KFold,StratifiedKFold
import gc
from sklearn.linear_model import LogisticRegression
from bayes_opt import BayesianOptimization
```

```python
n_folds=5
folds=KFold(n_splits=n_folds,shuffle=True,random_state=100)
feats=[f for f in train.columns if f not in ['Id','Cover_Type','Soil_Type7','Soil_Type15']]
```

```python
y=train['Cover_Type']
del train['Cover_Type']
```

```python
test_df=test.drop(['Id','Soil_Type7','Soil_Type15'],axis=1)
```

### LGBM

```python
oof_preds = np.zeros(train.shape[0])
sub_preds = np.zeros((test.shape[0],7))
    
feature_importance_df = pd.DataFrame()
```

### Bayesian Optimisation

```python
bayesian_tr_index, bayesian_val_index  = list(StratifiedKFold(n_splits=5, shuffle=True, random_state=1).split(train.values, y.values))[0]
```

```python
def LGB_bayesian(
    num_leaves,  # int
    min_data_in_leaf,  # int
    learning_rate,
    lambda_l1,
    feature_fraction,
    max_depth):
    
    
    
    # LightGBM expects next three parameters need to be integer. So we make them integer
    num_leaves = int(num_leaves)
    min_data_in_leaf = int(min_data_in_leaf)
    max_depth = int(max_depth)

    assert type(num_leaves) == int
    assert type(min_data_in_leaf) == int
    assert type(max_depth) == int

    param = {
        'num_leaves': num_leaves,
        'min_data_in_leaf': min_data_in_leaf,
        'learning_rate': learning_rate,
        'max_depth': max_depth,
        'lambda_l1': lambda_l1,
        'num_class':7,
        'seed': 123,
        'bagging_seed': 123,
        'drop_seed': 123,
        'data_random_seed': 123,
        'objective': 'multiclass',
        'boosting_type': 'gbdt',
        'verbose': 1,
        'metric': 'multi_logloss',  

    }    
    
    
    X_train, X_valid = train.iloc[bayesian_tr_index][feats], train.iloc[bayesian_val_index][feats]
    y_train, y_valid = y.iloc[bayesian_tr_index], y.iloc[bayesian_val_index]
    model=lgb.LGBMClassifier(**param,n_estimators=10000,n_jobs=-1)
    num_round = 10000
    clf=model.fit(X_train,y_train,
                    eval_set = [(X_train,y_train),(X_valid,y_valid)],eval_metric='multi_logloss',
                    verbose=100,
                    early_stopping_rounds = 200)
    
    
    predictions = model.predict(train.iloc[bayesian_val_index][feats].values)   
    
    score = model.score(X_valid,y_valid)
    
    return score
```

```python
bay_param={
    'num_leaves': (25,70), 
    'min_data_in_leaf': (10,70),  
    'learning_rate': (0.01, 0.3),
    'feature_fraction': (0.05, 0.5),
    'lambda_l1': (0, 5.0), 
    'max_depth':(3,15)
}
```

```python
LGB_BO = BayesianOptimization(LGB_bayesian, bay_param, random_state=123)
```

```python
init_points=3
n_iter=3
```

```python
print('-' * 150)

with warnings.catch_warnings():
    warnings.filterwarnings('ignore')
    LGB_BO.maximize(init_points=init_points,n_iter=n_iter,alpha=1e-06)
```

```python
LGB_BO.max['params']
```

```python
param = {'num_leaves': 25,
         'objective':'multiclass',
         'max_depth': 13,
         'num_class':7,
         'learning_rate': 0.05,
         "boosting": "gbdt",
         "feature_fraction": 0.8,
         "bagging_freq": 1,
         "bagging_fraction": 0.8 ,
         "bagging_seed": 11,
         "metric": ['multi_logloss'],
         'is_training_metric':True,
         "lambda_l1": 4,
         "random_state": 133,
         "verbosity": -1}
```

```python
score=0
for n_folds,(train_idx,valid_idx) in enumerate(folds.split(train.values,y.values)):
    print("fold n{}".format(n_folds+1))
    X_train, X_valid = train.iloc[train_idx][feats], train.iloc[valid_idx][feats]
    y_train, y_valid = y.iloc[train_idx], y.iloc[valid_idx]
    model=lgb.LGBMClassifier(**param,n_estimators=10000,n_jobs=-1)
    num_round = 10000
    clf=model.fit(X_train,y_train,
                    eval_set = [(X_train,y_train),(X_valid,y_valid)],eval_metric='multi_logloss',
                    verbose=100,
                    early_stopping_rounds = 200)
    
    #clf.fit(train_x,train_y,eval_set=[(train_x,train_y),(valid_x,valid_y)],verbose=500,eval_metric="auc",early_stopping_rounds=100)
    
    oof_preds[valid_idx]=model.predict(X_valid)
    sub_preds+=model.predict_proba(test_df)/folds.n_splits
    score+=model.score(X_valid,y_valid)
    fold_importance_df=pd.DataFrame()
    fold_importance_df['features']=feats
    fold_importance_df['importance']=model.feature_importances_
    fold_importance_df['folds']=n_folds+1
    print(f'Fold {n_folds+1}: Most important features are:\n')
    for i in np.argsort(fold_importance_df['importance'])[-5:]:
        print(f'{fold_importance_df.iloc[i,0]}-->{fold_importance_df.iloc[i,1]}')
    
    feature_importance_df=pd.concat([feature_importance_df,fold_importance_df],axis=0)
    print('Fold: {} score: {}'.format(n_folds,clf.score(X_valid,y_valid)))
    #print('Fold %2d Accuracy : %.6f' % (n_folds + 1, accuracy_score(y.iloc[valid_idx], oof_preds[valid_idx])))
    #del clf
    gc.collect()
    


#print('Full categorisation accuracy score %.6f' % (accuracy_score(y,oof_preds)))
print('Avg Accuracy', score / folds.n_splits)
test['Cover_Type']=sub_preds
              
```

```python
sample_submission['Cover_Type']=sub_preds.argmax(axis=1)+1
```
